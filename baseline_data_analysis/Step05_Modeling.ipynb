### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('forkedBigdataAnalystCert/part3/ch6/pimaIndiansdiabetes2.csv')
df
```

```python
df.info()
```

```python
df = df.dropna().reset_index(drop = True)
df
```

```python
X = df.drop(columns = 'diabetes')
y = LabelEncoder().fit_transform(df['diabetes'])#pos/neg를 1/0으로 변경 
```

```python
Xtr, Xte, ytr, yte = train_test_split(X, y, random_state = SEED, test_size = 0.2)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(Xtr, ytr)
pred = lr.predict(Xte)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
```

```python
confusion_matrix(yte, pred, labels = [1, 0])
```

```python
accuracy_score(yte, pred)
```

```python
lrlasso = LogisticRegression(penalty='l1', solver = 'saga', max_iter = 10000)
lrlasso.fit(Xtr, ytr)
lrlasso
```

## RandomForest

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
df = pd.read_csv('forkedBigdataAnalystCert/part3/ch6/pimaIndiansdiabetes2.csv')
df = df.dropna().reset_index(drop = True)
df
```

```python
X = df.drop(columns = ['diabetes'])
y = df['diabetes']
y = LabelEncoder().fit_transform(y)
```

```python
Xtr, Xte, ytr, yte = train_test_split(X, y, random_state=SEED)
rfclf = RandomForestClassifier()
rfclf.fit(Xtr, ytr)
pred = rfclf.predict(Xte)
accuracy_score(yte, pred)
```

## SVM Classifier

```python
from sklearn.svm import SVC
```

```python
df = pd.read_csv('forkedBigdataAnalystCert/part3/ch6/pimaIndiansdiabetes2.csv')
df = df.dropna().reset_index(drop = True)
df
```

```python
X = df.drop(columns = ['diabetes'])
y = df['diabetes']
y = LabelEncoder().fit_transform(y)
```

```python
Xtr, Xte, ytr, yte = train_test_split(X, y, random_state = SEED, test_size = 0.2)
```

## Data types

원래 '자료형'은 int64, float64, datetime64 등 다양하게 있지만, 데이터 분석 과정에서 사용되는 자료형은 정수/실수 등이 중요한 것이 아니다. 자료형이 '문자열'이어도 실제 자료에서는 key의 역할을 할 수도 있고, binary data를 나타낼 수도 있다.

아래는 칼럼별로 '자료형'을 나타내는 함수이다.

```python
def group_columns_by_dtype(df) -> dict:

    dtype_dict = {}
    
    for col in df.columns:
        dtype = df[col].dtype
        if dtype not in dtype_dict:
            dtype_dict[dtype] = []
        dtype_dict[dtype].append(col)
    
    return dtype_dict
```

실제 환경에서, 데이터의 분류는 다음과 같다.

* 명목형 : 이항분포, 다항분포 등을 따르는 데이터. 이항분포를 따르는 경우 이산형 데이터라고도 함.
* 순서형 : 명목형 변수 중, 순서가 있는 것.
* count data : 보통 int 형태. 실수형처럼 취급할 수도 있지만, zero-inflated 등 특성도 있음.
* 연속형 : 주로 실수. 절대영점이니 상대영점이니 하는 얘기가 있지만... 그게 문제됐던 기억은 없음.

## `pd.merge()` : key를 기준으로 데이터 합치기

데이터프레임 병합 방법은 크게 다음의 네 가지가 있다.
* inner join : `pd.merge(left_df, right_df, on = 'keycolumn')`. 열이 두 데이터프레임에 모두 있어야 반환
* outer join : `pd.merge(left_df, right_df, on = 'keycolumn', how = 'outer')`. 열이 두 데이터프레임에 하나라도 있으면 반환
* left/right join : `pd.merge(left_df, right_df, on = 'keycolumn', how = 'left'/'right')`. 열이 왼쪽/오른쪽 데이터프레임에 있는 것만 반환

```python
pd.merge(df2, df4, on='A', how='inner')
```

두 데이터프레임에서 키가 서로 다른 경우, `left_on`과 `right_on`을 각각 사용해 키를 선언한다.

```python
pd.merge(df2, df4, left_on='B', right_on='B', suffixes=['_df2', '_df4'])
```

`merge`의 경우, pd.DataFrame의 메소드로도 존재하여 다음과 같이 사용할 수도 있다.

```python
df2.merge(df4, left_on='B', right_on='B')
```

기본적인 사용법은 `pd.DataFrame.merge()`와 거의 똑같다. 단, 기본 join 방법이 left join이다.

```python
df2.join(df3, lsuffix='_df2', rsuffix='_df3', how='inner')
```

```python
df2.join(df3, lsuffix='_df2', rsuffix='_df3')
```

# Data Wrangling : grouped operation

그룹 연산의 개념 : 분할(split) -> 적용(apply) -> 결합(combine)

![grouped operation](figures/split-apply-combine.png)

이미지의 소스코드는 https://jakevdp.github.io/PythonDataScienceHandbook/06.00-figure-code.html#Split-Apply-Combine 에서 확인할 수 있다.

```python
df = sns.load_dataset('penguins')
df
```

### split : `group_by()`

```python
df_grouped = df.groupby(['species'])
df_grouped.head()
```

현재는 그룹화만 이루어졌으므로 큰 차이가 나지 않는다. 그러나 분할 여부를 아래와 같이 확인 가능하다.

```python
for k, v in df_grouped:
    print(k)
    display(v.head(2))
```

여러 열을 기준으로 그룹화 역시 가능하다.

```python
df_multigrouped = df.groupby(['species', 'island'])
df_multigrouped.head()
```

### apply and combine : methods for aggregation

```python
df_grouped['bill_length_mm'].mean()
```

```python
df_multigrouped[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].mean()
```

다음과 같은 집계 함수가 존재한다.
* `.count()`, `.size()` : NAs 제외 데이터 수/NAs 포함 데이터 수
* `.describe()`
* `.mean()`, `.std()`, `.var()`, `.sum()`
* `.min()`, `.max()`, `.quantile(q=0.4)`, `.median()`
* `.first()`, `last()`, `.nth(n=10)`

직접 함수를 만들어서 입력하고 싶다면, `.agg()`와 `.transform()`이라는 메소드를 사용할 수 있다.

```python
def myrange(x:pd.DataFrame) -> pd.Series:
    return x.max() - x.min()

def myzscore(x:pd.Series) -> pd.Series:
    return (x - x.mean()) / x.std()
```

`.agg()` 함수는 집계 함수에 따라, 결합된 결과를 반환한다.

```python
df.groupby(['species', 'island'])[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].agg(myrange)
```

```python
df.groupby(['species', 'island']).agg({'bill_length_mm': ['min', 'max'], 'bill_depth_mm': ['sum']})
```

참고로, 이 경우 series의 methods에 `.myrange()`가 없으므로 오류가 발생한다.

`.transform()` 함수는 그룹별로 함수를 적용하되, 그 결과를 원본 데이터프레임과 같은 형태로 재변환해 결과를 return한다. 즉 각 그룹별로 계산 + 원래의 데이터프레임의 각 row마다 결과 할당을 동시에 한다.

```python
df.groupby(['species', 'island'])[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].transform('mean')
```

```python
df.groupby(['species', 'island'])[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].transform(myzscore)
```

```python
df.groupby(['species'])[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].transform(myzscore).reset_index(drop=True)
```

`.apply()` 함수를 사용하여 각 그룹별로 함수를 적용하는 방법도 있다. 위와 결과가 동일하다.

```python
df.groupby(['species', 'island'])[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].apply(myzscore)
```

`.filter()`를 통해 조건에 해당하는 그룹에 포함된 row만을 반환받을 수도 있다.

```python
df.groupby(['species', 'island']).filter(lambda x: x['body_mass_g'].mean() < 4000)
```

```python
df.groupby(['species', 'island']).filter(lambda x: x['body_mass_g'].mean() < 4000 and x['bill_length_mm'].mean() < 40)
```

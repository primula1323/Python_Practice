# Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.model_selection import train_test_split
```

```python
learning_cols = ['Item_Weight', 'Item_Fat_Content', 'Item_Visibility', 'Item_Type', 'Item_MRP', 'Outlet_Identifier', 'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']
target_cols = ['Item_Outlet_Sales']

obj_cols = list(df_train.columns[df_train.dtypes == object])
obj_cols.remove('Item_Identifier')

real_cols = [col for col in learning_cols if col not in obj_cols]
```

```python
encoder = OrdinalEncoder()
encoder.fit(df_train[obj_cols])
df_train[obj_cols] = encoder.fit_transform(df_train[obj_cols])
df_test[obj_cols] = encoder.fit_transform(df_test[obj_cols])

scaler = StandardScaler()
scaler.fit(df_train[real_cols])
df_train[real_cols] = scaler.fit_transform(df_train[real_cols])
df_test[real_cols] = scaler.fit_transform(df_test[real_cols])
```

```python
imputer_Outlet_Size = SimpleImputer(strategy = 'most_frequent')
imputer_Outlet_Size.fit(df_train[obj_cols])
df_train[obj_cols] = imputer_Outlet_Size.fit_transform(df_train[obj_cols])

imputer_Item_Weight = KNNImputer()
imputer_Item_Weight.fit(df_train[learning_cols])
df_train[learning_cols] = imputer_Item_Weight.fit_transform(df_train[learning_cols])
df_test[learning_cols] = imputer_Item_Weight.fit_transform(df_test[learning_cols])
```

```python
X_train, X_val, y_train, y_val = train_test_split(df_train[learning_cols], df_train[target_cols], test_size = 0.2, random_state = 42)
```

# Learning

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
Model_RF1 = RandomForestRegressor(random_state = 42)
Model_RF1.fit(X_train, y_train)
```

```python
pred = Model_RF1.predict(X_val)
rmse1 = root_mean_squared_error(y_val, pred)
rmse1
```

```python
Model_xgb1 = XGBRegressor(random_state = 42)
Model_xgb1.fit(X_train, y_train)
```

```python
pred = Model_xgb1.predict(X_val)
rmse2 = root_mean_squared_error(y_val, pred)
rmse2
```

```python
final_model = Model_RF1
```

# File make

```python
predicted_value = final_model.predict(df_test[learning_cols])
submit = pd.DataFrame({'pred':predicted_value})
submit.to_csv('result.csv', index = False)
```

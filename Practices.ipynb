## 01

```python
import numpy as np, pandas as pd
df = pd.read_csv('Data/P210204-01.csv')
```

```python
df.info()
```

```python
df.rename(columns = {'Reached.on.Time_Y.N':'Target'}, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as RFclf
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop(columns = ['Target', 'ID'])
y = df['Target']
```

```python
Xtr, Xte, ytr, yte = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

```python
len(Xtr)
```

```python
colnames = list(X.columns)
colnames
```

```python
colnames.remove('Warehouse_block')
colnames.remove('Mode_of_Shipment')
colnames.remove('Product_importance')
colnames.remove('Gender')
colnames
```

```python
catcols = ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']
X[catcols]
```

```python
Xcon = Xtr[colnames]
scaler = MinMaxScaler()
scaler.fit(Xcon)
Xcon = pd.DataFrame(scaler.fit_transform(Xcon), columns = colnames)
Xcon
```

```python
Xcat = pd.get_dummies(Xtr[catcols], drop_first = True).reset_index(drop=True)
```

```python
Xtrs = pd.concat([Xcon, Xcat], axis = 1)
Xtrs
```

```python
Xtecon = Xte[colnames]
Xtecon = pd.DataFrame(scaler.fit_transform(Xtecon), columns = colnames)
Xtecat = pd.get_dummies(Xte[catcols], drop_first = True).reset_index(drop=True)
Xte = pd.concat([Xtecon, Xtecat], axis = 1)
Xte
```

```python
ytr.reset_index(drop = True, inplace = True)
yte.reset_index(drop = True, inplace = True)
```

```python
clf = RFclf()
clf.fit(Xtrs, ytr)
```

```python
pred = clf.predict_proba(Xte)
```

```python
pred_prob = [pred[i][1] for i in range(len(pred))]
pred_prob
```

```python
predf = pd.read_csv('Data/P210204-02.csv')
predf
```

```python
predcat = pd.get_dummies(predf[catcols], drop_first=True).reset_index(drop = True)
```

```python
predcon = pd.DataFrame(scaler.fit_transform(predf[colnames]), columns = colnames)
```

```python
predf = pd.concat([predcon, predcat, predf['ID']], axis = 1)
```

```python
pred = clf.predict_proba(predf.drop(columns = ['ID']))
```

```python
final_pred = pd.concat([predf['ID'], pd.DataFrame([pred[i][1] for i in range(len(predf))], columns = ['result'])], axis = 1)
```

```python
final_pred
```

```python
final_pred.to_csv('Answers/0201.csv', index = False)
```
